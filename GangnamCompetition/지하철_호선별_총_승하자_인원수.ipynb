```python
import pandas as pd

def load_csv(file_path):

    #CSV 파일을 읽어서 데이터프레임으로 반환하는 함수
    #:param file_path: 읽어올 CSV 파일의 경로
    #:return: 데이터프레임

    try:
        # CSV 파일 읽기
        subway = pd.read_csv(file_path, encoding='cp949')
        return subway
    except Exception as e:
        print("Error:", e)
        return None

# CSV 파일 경로 설정
csv_file_path = "서울시 지하철 호선별 역별 총 인원.csv"  # 실제 파일 경로로 변경 필요

# CSV 파일 불러오기
subway = load_csv(csv_file_path)

# 데이터프레임 확인
subway
```

```python
subway[["역명"]]
```

```python
subway[subway["호선명"] == "2호선"][["호선명","역명", "승차총승객수", "하차총승객수"]]
```

```python
# 원하는 호선명 설정
desired_lines = ["2호선", "7호선", "9호선", "3호선", "분당선"]

# 원하는 역명 설정
desired_station = ["강남구청", "강남", "개포동", "구룡", "논현", "대모산입구", "대청", "대치", "도곡", "매봉",
                   "봉은사", "삼성", "삼성중앙", "선릉", "선정릉", "수서", "신논현", "신사", "압구정로데오",
                   "압구정", "양재", "언주", "역삼", "일원", "청담", "학동", "학여울", "한티"]

# 해당 호선명과 역명에 해당하는 행 선택
passenger_sum = subway[subway["호선명"].isin(desired_lines) & subway["역명"].isin(desired_station)]

# 승차총승객수와 하차총승객수를 더하여 새로운 열 추가
passenger_sum["총승객수"] = passenger_sum["승차총승객수"] + passenger_sum["하차총승객수"]

# 필요한 열만 선택하여 새로운 데이터프레임 생성
passenger_df = passenger_sum[["호선명", "역명", "총승객수"]]

# 결과 확인
passenger_df
```

```python
# 총승객수를 기준으로 내림차순으로 정렬
passenger_df_sorted = passenger_df.sort_values(by='총승객수', ascending=False)

# 상위 5개의 데이터 선택
top_5_passenger = passenger_df_sorted.head(5)

# 결과 확인
top_5_passenger
```

# 한달 동안 승하자가 제일 많은 호선명과 역명

```python
# 겹치는 호선명과 역명을 그룹화하여 총승객수를 합산
grouped_passenger = passenger_df.groupby(["호선명", "역명"]).sum().reset_index()

# 총승객수를 기준으로 내림차순으로 정렬
grouped_passenger_sorted = grouped_passenger.sort_values(by='총승객수', ascending=False)

# 상위 5개의 데이터 선택
top_5_passenger = grouped_passenger_sorted.head(5)

# 결과 확인
top_5_passenger
```
